```python
def replacer(df):
    import pandas as pd
    Q = pd.DataFrame(df.isna().sum(),columns=["CT"])
    W = Q[Q.CT > 0].index
    for i in W:
        if(df[i].dtypes == "object"):
            x = df[i].mode()[0]
            df[i] = df[i].fillna(x)
        else:
            x = df[i].mean()
            df[i] = df[i].fillna(x)
```

```python
replacer(data)
```

```python
Y = data[["SalePrice"]]
X = data.drop(labels=["SalePrice"],axis=1)
```

```python
X = X.drop(['Id'],axis=1)
```

```python
def data_prep(X):
    cat = []
    con = []
    for i in X.columns:
        if(X[i].dtypes == "object"):
            cat.append(i)
        else:
            con.append(i)
            
    import pandas as pd
    X1 = pd.get_dummies(X[cat])
    from sklearn.preprocessing import StandardScaler
    ss = StandardScaler()
    X2 = pd.DataFrame(ss.fit_transform(X[con]),columns=con)
    X3 = X2.join(X1)
    return X3
```

```python
Xnew = data_prep(X)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=31)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
Xnew.shape
```

```python
nn = Sequential()
nn.add(Dense(100,activation="relu",input_dim=(288)))
nn.add(Dense(100,activation="relu"))
nn.add(Dense(1))
```

```python
nn.compile(loss="mean_squared_error")
nn.fit(xtrain,ytrain,epochs=50)
```

```python
predicated_price=nn.predict(xtest)
```
